3. You have a dataset named stock_prices.csv that contains daily stock prices of a company. The dataset has columns date and closing_price. Write a Python script to read the dataset, convert the date column to a datetime object, and then perform a time series analysis by plotting the closing price over time with a rolling 30-day moving average.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Prateek\Downloads\stock_prices.csv")

df
```

```python
df['date']= pd.to_datetime(df['date'])
```

```python
df.set_index('date', inplace=True)
```

```python
df['30-Day Avg'] = df['closing_price'].rolling(window=30).mean()
```

```python
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df.index, df['closing_price'], label='closing_price', alpha=0.7)
plt.plot(df.index, df['30-Day Avg'], label='30-Day Avg', linewidth=2)
plt.title('Closing Price with 30-Day Moving Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```
